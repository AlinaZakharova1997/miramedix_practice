```
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'

def load_model(model_name):
  tokenizer = MarianTokenizer.from_pretrained(model_name)
  model = MarianMTModel.from_pretrained(model_name).to(DEVICE)
  return tokenizer, model
```

Greedy search

```
source = []
hypothesis = []
reference = []
compare = {}
```

```
for i in range(len(target)):

  target_item = target[i] # reference
  #print(target_item, 'TARGET')
  src_item = src[i] # source
  #print(src_item, 'src_item')
  inputs = tokenizer.encode(src_item, return_tensors="pt", max_length=512, truncation=True)
  #print(inputs, 'inputs')
  greedy_outputs = model.generate(inputs) 
  result = tokenizer.decode(greedy_outputs[0], skip_special_tokens=True) # hypothesis
  compare[target_item] = result
  #print(result, 'RESULT')

  # collecting data for metrics count
  source.append(src_item)
  hypothesis.append(result)
  reference.append([target_item])
  
```

```
beam_source = []
beam_hypothesis = []
beam_reference = []
beam_compare = {}
```

```
for i in range(len(target)):

  target_item = target[i] # reference
  #print(target_item, 'TARGET')
  src_item = src[i] # source
  #print(src_item, 'src_item')
  inputs = tokenizer.encode(src_item, return_tensors="pt", max_length=512, truncation=True)
  #print(inputs, 'inputs')
  
  beam_outputs = model.generate(
    inputs.cuda(), 
    num_beams=5, 
    num_return_sequences=5,
    early_stopping=True,
)
  result = tokenizer.decode(beam_outputs[0], skip_special_tokens=True) # hypothesis
  beam_compare[target_item] = result
  #print(result, 'RESULT')

  # collecting data for metrics count
  beam_source.append(src_item)
  beam_hypothesis.append(result)
  beam_reference.append([target_item])
  
```

```
# We need this to prepare outputs for bleu metric

def ref_to_bleu(beam_reference):
  beam_ref = []
  for item in beam_reference:
    text = item[0]
    tokenized = tokenizer.tokenize(text)
    beam_ref.append([tokenized])
  return beam_ref

def hyp_to_bleu(beam_hypothesis):
  beam_hyp = []
  for text in beam_hypothesis:
    tokenized = tokenizer.tokenize(text)
    beam_hyp.append(tokenized)
  return beam_hyp
```
