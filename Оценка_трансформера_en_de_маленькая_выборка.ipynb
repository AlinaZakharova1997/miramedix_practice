Model we will test

```
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'

def load_model(model_name):
  tokenizer = MarianTokenizer.from_pretrained(model_name)
  model = MarianMTModel.from_pretrained(model_name).to(DEVICE)
  return tokenizer, model
```

Тут будут вспомогательные функции для ускорения работы

Эти функции нужны для преобразования результатов для вычисления метрики bleu

```
# We need this to prepare outputs for bleu metric

def ref_to_bleu(beam_reference):
  beam_ref = []
  for item in beam_reference:
    text = item[0]
    tokenized = tokenizer.tokenize(text)
    beam_ref.append([tokenized])
  return beam_ref

def hyp_to_bleu(beam_hypothesis):
  beam_hyp = []
  for text in beam_hypothesis:
    tokenized = tokenizer.tokenize(text)
    beam_hyp.append(tokenized)
  return beam_hyp
```

Эти функции нужны для преобразования данных для вычисление метрики comet

```
def comet_data_maker(source, hypothesis, reference):
  comet_reference = []
  for d in reference:
    comet_reference.append(d[0])
  comet_data = [{"src": src, "mt": mt, "ref": ref} for src, mt, ref in
                  zip(source, hypothesis, comet_reference)]
  return comet_data
```

Эта функция будет вычислять метрики, которые мы выбрали для оценки модели
Мы взяли BLEU, ROUGE, SACREBLEU, METEOR, COMET 

```
BLEU = 'bleu'
ROUGE = 'rouge'
SACREBLEU = 'sacrebleu'
METEOR = 'meteor'
COMET = 'comet'
```

```
def compute_metric(metric_name, references, predictions):
    return load_metric(metric_name).compute(predictions=predictions,
                                            references=references)
```

Эта функция для того, чтобы посчитать все нужные метрики и записать их в отдельный json file

```
def get_results(filename, source, hypothesis, reference):

  path = '/content/drive/MyDrive/'+filename
  comet_data = comet_data_maker(source, hypothesis, reference)
  bleu_references = ref_to_bleu(reference)
  bleu_predictions = hyp_to_bleu(hypothesis)

  results = {BLEU: compute_metric(BLEU, bleu_references, bleu_predictions),
               ROUGE: compute_metric(ROUGE, reference, hypothesis),
               SACREBLEU: compute_metric(SACREBLEU, reference,
                                         hypothesis),
               METEOR: compute_metric(METEOR, reference, hypothesis),
               COMET: comet_model.predict(comet_data, gpus=0),
               }
  with open(path, 'w', encoding='utf8') as f:
    json.dump(results, f)
  return results
```

Greedy search

Это просто самый обычный greedy search, тут из параметров только максимальная длина, которая собств максимальная 512, чтобы не точно резать предложения.

```
source = []
hypothesis = []
reference = []
compare = {}
```

```
for i in range(len(target)):

  target_item = target[i] # reference
  #print(target_item, 'TARGET')
  src_item = src[i] # source
  #print(src_item, 'src_item')
  inputs = tokenizer.encode(src_item, return_tensors="pt", max_length=512, truncation=True)
  #print(inputs, 'inputs')
  greedy_outputs = model.generate(inputs.cuda()) 
  result = tokenizer.decode(greedy_outputs[0], skip_special_tokens=True) # hypothesis
  compare[target_item] = result
  #print(result, 'RESULT')

  # collecting data for metrics count
  source.append(src_item)
  hypothesis.append(result)
  reference.append([target_item])
  
```
