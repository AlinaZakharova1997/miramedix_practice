Получается у меня N зависимых выборок с количественными прикнаками. 

Количесвенный признак = числовое значение метрики комета для предложения.

Что мне надо с этим делать?

Мне надо посмотреть, отличаются ли эти выборки друг от друга и являются ли эти различия статистически значимыми.

Если распеделения нормальные, то можно использовтаь дисперсионный анализ, иначе - Непараметрический дисперсионный анализ.
Получается сначала надо понять, как проверить на нормальность.

```python
from scipy import stats
```

```python
for comet in all_data:
  stat, p = stats.normaltest(comet['COMET_LIST']) # Критерий согласия Пирсона
  print('Statistics=%.3f, p-value=%.3f' % (stat, p))
  alpha = 0.05
  if p > alpha:
    print('Принять гипотезу о нормальности')
  else:
    print('Отклонить гипотезу о нормальности')
```

Если распеределения ненормальные, значит ищем критерии для 3-х и более ненормальных распределений при условии что выборки зависимые
Нам может подойти тест Фридмана

Предположение по умолчанию или нулевая гипотеза заключается в том, что несколько парных выборок имеют одинаковое распределение. Отказ от нулевой гипотезы указывает на то, что один или несколько парных образцов имеют другое распределение.

    Не в состоянии отклонить H0: Парные выборки распределены одинаково.
    Отклонить H0: Парные выборки распределений не равны.

Тест предполагает две или более парных выборок данных с 10 или более выборками на группу.

Тест Фридмана - это непараметрическая статистическая процедура для сравнения более двух связанных выборок. Параметрическим эквивалентом этого теста является дисперсионный анализ повторных измерений (ANOVA). Когда тест Фридмана приводит к значительным результатам, по крайней мере один из образцов отличается от других образцов.

```python
stat, p = stats.friedmanchisquare(comet1['COMET_LIST'], comet2['COMET_LIST'], comet3['COMET_LIST'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
  print('Принять гипотезу о нормальности')
else:
  print('Отклонить гипотезу о нормальности')
```

Если я хочу посмотреть разницу между двумя выборками, то тут мне с учетом всех факторов подходит Вилксон

```python
stat, p = stats.wilcoxon(comet1['COMET_LIST'], comet2['COMET_LIST'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
  print('Принять гипотезу о нормальности')
else:
  print('Отклонить гипотезу о нормальности')
```
